```python
# Import necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Specify the path to the CSV files

file_path1 = r'C:\Users\BINDU\Downloads\Hospital_inpatient_discharges.csv'
file_path2 = r'C:\Users\BINDU\Downloads\NYC_EH_DataPortal.csv'
file_path3 = r'C:\Users\BINDU\Downloads\Air_Quality.csv'
```

```python
# Load the three datasets (assuming they are CSV files)
hospital_data = pd.read_csv(file_path1)
health_data = pd.read_csv(file_path2)
air_quality_data = pd.read_csv(file_path3)
```

```python
# Display the first few rows of each dataset to understand their structure
print("Hospital Data - First 5 rows:")
display(hospital_data.head())

print("Environmental Health Data - First 5 rows:")
display(health_data.head())

print("Air Quality Data - First 5 rows:")
display(air_quality_data.head())
```

```python
#Data Exploration 

# Information about dataset and summary statistics
print("Hospital Data Info:")
hospital_data.info()
print("Hospital Data Summary Statistics:")
display(hospital_data.describe())

print("Health Data Info:")
health_data.info()
print("Environmental Health Data Summary Statistics:")
display(health_data.describe())

print("Air Quality Data Info:")
display(air_quality_data.info())

print("Air Quality Data Summary Statistics:")
display(air_quality_data.describe())
```

```python
#Handling Missing Values

print("Hospital Data - Missing Values:")
display(hospital_data.isnull().sum())

print("Environmental Health Data - Missing Values:")
display(health_data.isnull().sum())

print("Air Quality Data - Missing Values:")
display(air_quality_data.isnull().sum())
```

```python
# Handling missing values - Drop rows with missing 'Total Charges' or 'Total Costs'
hospital_data = hospital_data.dropna(subset=['Total Charges', 'Total Costs'])

# For air_quality_data, we can impute missing 'Data Value' with the median
air_quality_data['Data Value'] = air_quality_data['Data Value'].fillna(air_quality_data['Data Value'].median())

# For health_data, fill missing 'Age-adjusted rate' with the median
health_data.columns = health_data.columns.str.strip()
health_data['Age-adjusted rate per 10000'] = health_data['Age-adjusted rate per 10000'].fillna(
    health_data['Age-adjusted rate per 10000'].median())
```

```python
air_quality_data = air_quality_data.drop(columns=['Message'])
```

```python
# Drop columns with more than 50% missing values
hospital_data = hospital_data.drop(columns=['Payment Typology 2', 'Payment Typology 3', 'Birth Weight'])
# Drop rows with missing values in critical identifier columns
hospital_data = hospital_data.dropna(subset=['Hospital Service Area', 'Hospital County', 'Operating Certificate Number', 'Permanent Facility Id'])
# Impute categorical columns with mode
hospital_data['APR Severity of Illness Description'].fillna(hospital_data['APR Severity of Illness Description'].mode()[0], inplace=True)
hospital_data['APR Risk of Mortality'].fillna(hospital_data['APR Risk of Mortality'].mode()[0], inplace=True)
# Fill missing ZIP codes with mode, or use another location-based logic if possible
hospital_data['Zip Code - 3 digits'].fillna(hospital_data['Zip Code - 3 digits'].mode()[0], inplace=True)
# Fill missing values in CCSR columns with "Not Available"
hospital_data['CCSR Procedure Code'].fillna("Not Available", inplace=True)
hospital_data['CCSR Procedure Description'].fillna("Not Available", inplace=True)
```

```python
# Confirm that the missing values are handled
print("Hospital Data - Missing Values After Handling:")
display(hospital_data.isnull().sum())

print("Environmental Health Data - Missing Values After Handling:")
display(health_data.isnull().sum())

print("Air Quality Data - Missing Values After Handling:")
display(air_quality_data.isnull().sum())
print(health_data.columns)
```

```python
# Finding Duplicate Values

#Check for duplicate rows in each dataset
print("Hospital Data - Duplicates:")
hospital_data_duplicates = hospital_data.duplicated().sum()

print("Environmental Health Data - Duplicates:")
health_data_duplicates = health_data.duplicated().sum()

print("Air Quality Data - Duplicates:")
air_quality_data_duplicates = air_quality_data.duplicated().sum()

# Remove duplicates if found
if hospital_data_duplicates > 0:
    hospital_data = hospital_data.drop_duplicates()

if health_data_duplicates > 0:
    health_data = health_data.drop_duplicates()

if air_quality_data_duplicates > 0:
    air_quality_data = air_quality_data.drop_duplicates()
```

```python
# Confirm no duplicates are present
print(f"Hospital Data - Duplicates After Removal: {hospital_data.duplicated().sum()}")
print(f"Environmental Health Data - Duplicates After Removal: {health_data.duplicated().sum()}")
print(f"Air Quality Data - Duplicates After Removal: {air_quality_data.duplicated().sum()}")
```

```python
#Data type Conversion
# Convert 'Discharge Year' to categorical
hospital_data['Discharge Year'] = hospital_data['Discharge Year'].astype('category')

# Convert 'Start_Date' to datetime
air_quality_data['Start_Date'] = pd.to_datetime(air_quality_data['Start_Date'], errors='coerce')

# Convert 'GeoID' in health_data to categorical
health_data['GeoID'] = health_data['GeoID'].astype('category')

# Check data types after conversion
print("Hospital Data - Data Types After Conversion:")
print(hospital_data.dtypes)

print("Environmental Health Data - Data Types After Conversion:")
print(health_data.dtypes)

print("Air Quality Data - Data Types After Conversion:")
print(air_quality_data.dtypes)
```

```python
#Outlier Detection and Handling

# Visualize outliers for 'Total Charges' in hospital_data using boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x=hospital_data['Total Charges'])
plt.title('Boxplot of Total Charges')
plt.show()
```

```python
# Handle outliers - Cap 'Total Charges' at the 99th percentile
upper_limit = hospital_data['Total Charges'].quantile(0.99)
hospital_data = hospital_data[hospital_data['Total Charges'] <= upper_limit]
```

```python
# Re-check for outliers after handling
plt.figure(figsize=(12, 6))
sns.boxplot(x=hospital_data['Total Charges'])
plt.title('Boxplot of Total Charges')
plt.show()
```

```python
#Data Analysis
#Hospital Data Analysis: Hospital discharges by gender and age group

# Group hospital data by 'Gender' and 'Age Group' to analyze discharges by these categories
hospital_discharge_gender_age = hospital_data.groupby(['Gender', 'Age Group']).size().unstack()

# Display the summary
hospital_discharge_gender_age

# Total discharges by gender (sum across all age groups)
total_discharges_gender = hospital_discharge_gender_age.sum(axis=1)
print("Total discharges by gender:")
print(total_discharges_gender)

# Total discharges by age group (sum across all genders)
total_discharges_age = hospital_discharge_gender_age.sum(axis=0)
print("\nTotal discharges by age group:")
print(total_discharges_age)

# Overall total discharges
total_discharges_all = hospital_discharge_gender_age.sum().sum()
print(f"\nTotal discharges across all categories: {total_discharges_all}")
```

```python
#Data Visualization 1: Bar Chart â€“ Total Discharges by Gender and Age Group (Hospital Data)

# Group hospital data by 'Gender' and 'Age Group' to analyze discharges by these categories
hospital_discharge_gender_age = hospital_data.groupby(['Gender', 'Age Group']).size().unstack()
```

```python
# Create a bar plot to visualize the number of hospital discharges by gender and age group
hospital_discharge_gender_age.plot(kind='bar', stacked=True, figsize=(10, 6))

# Adding titles and labels
plt.title('Hospital Discharges by Gender and Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Discharges')
plt.xticks(rotation=45)
plt.legend(title='Gender', loc='upper right')
#Display the Plot
plt.tight_layout()
plt.show()
```

```python
# Data Visualization 2: Scatter Plot
# Scatter plot for NYC EH Data: Age-adjusted rate per 10,000 vs Estimated annual rate per 10,000

plt.figure(figsize=(10, 6))

# Plotting the scatter plot
sns.scatterplot(x='Age-adjusted rate per 10000', y='Estimated annual rate per 10000', data=health_data, alpha=0.6)

# Adding titles and labels
plt.title('Scatter Plot: Age-adjusted Rate vs Estimated Annual Rate per 10,000')
plt.xlabel('Age-adjusted Rate per 10,000')
plt.ylabel('Estimated Annual Rate per 10,000')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Data Visualization 3: Correlation Heatmap
# Step 1: Extract the year from 'Start_Date' if it is not already done
air_quality_data['Year'] = air_quality_data['Start_Date'].dt.year

# Step 2: Create a DataFrame with 'Data Value' and 'Year' columns
data_value_year_df = air_quality_data[['Year', 'Data Value']].dropna()  # Remove missing values

# Step 3: Compute the correlation matrix between 'Data Value' and 'Year'
correlation_matrix = data_value_year_df.corr()

# Step 4: Create the heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Adding titles and labels
plt.title('Correlation Heatmap between Data Value and Year')
plt.tight_layout()
plt.show()
```
